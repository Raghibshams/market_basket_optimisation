```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn import linear_model
import seaborn as sns
%matplotlib inline
```

```python
from apyori import apriori
```

```python
df=pd.read_csv('Market_Basket_Optimisation.csv',header=None)
```

```python
df.head(5)
```

```python
transaction=[]
for i in range(0,7501):
    transaction.append([str(df.values[i,j]) for j in range(0,20)])
    

 
```

```python
rules=apriori(transaction,min_support=.002,min_confidence=.3,min_fit=3,max_length=2)
```

```python
results=list(rules)
results
```

```python
for i in results:
    print(i)
    print(' ')
```
